```python
import pandas as pd
import numpy as np

Classification_df= pd.read_csv('Classification_Data.csv')

Clustering_df=pd.read_csv('Clustering_Data.csv')
Clustering_df.head(20)
```

```python
print(Clustering_df.shape)
```

```python
for col in Clustering_df.columns:
    print(col)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
print(Clustering_df.info())
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10)
kmeans.fit(Clustering_df)
print(kmeans.cluster_centers_)
```

```python
Clustering_df['Clusters'] = kmeans.labels_
Clustering_df.head(20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=10)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(Clustering_df, test_size=2)
train_x = train.drop('Clusters', axis = 1)
train_y = train['Clusters']

test_x = test.drop('Clusters', axis = 1)
test_y = test['Clusters']
```

```python
knn_clf.fit(train_x, train_y)
predictions = knn_clf.predict(test_x)

from sklearn.metrics import accuracy_score
print(accuracy_score(test_y, predictions))
```

```python
Classification_df['Clusters'] = knn_clf.predict(Classification_df)
```
